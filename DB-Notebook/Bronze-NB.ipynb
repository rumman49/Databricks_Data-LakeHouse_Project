## Reading and writing the Source Crm files into tables



```python
df = spark.read.option("header", "true").option("inferSchema", "true").csv("/Volumes/lakehouse/bronze/source_files/cust_info.csv")

# cust_info --
df.write.mode("overwrite").format("delta").saveAsTable("lakehouse.bronze.cust_info")
```
